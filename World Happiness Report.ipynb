```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
```

```python
# Load the dataset
data = pd.read_csv('2015.csv') 
```

```python
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(data['Economy (GDP per Capita)'], 
                     data['Health (Life Expectancy)'], 
                     data['Happiness Score'], 
                     c=data['Happiness Score'], 
                     cmap='viridis', 
                     s=100)

# Labels
ax.set_xlabel('Economy (GDP per Capita)')
ax.set_ylabel('Health (Life Expectancy)')
ax.set_zlabel('Happiness Score')
plt.title('3D Rotational Visualization of Happiness Factors')

# Animation function to rotate the plot
def update(num):
    ax.view_init(elev=num, azim=num + 90)

ani = FuncAnimation(fig, update, frames=range(0, 360, 5), interval=100)
plt.show()
```

```python
# Count the number of countries in each region
region_counts = data['Region'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Countries by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Prepare data for the line graph
top_countries = data.sort_values('Happiness Score', ascending=False).head(10)
plt.figure(figsize=(14, 7))
sns.lineplot(data=top_countries, x='Country', y='Happiness Score', marker='o', palette='Set2', linewidth=2.5)
plt.title('Happiness Score Trend of Top 10 Countries', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Happiness Score', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
data = pd.read_csv('2015.csv')
```

```python
# Filter for the top 10 happiest countries
top_countries = data.nlargest(10, 'Happiness Score')[['Country', 'Happiness Score']]
top_countries['Year'] = 2015  # Assign the year of the data
```

```python
# Prepare data for linear regression
years = np.array([2015, 2020, 2025, 2030]).reshape(-1, 1)  # Years for prediction
predictions = []

# Fit a linear regression model for each country
for country in top_countries['Country']:
    country_data = top_countries[top_countries['Country'] == country]
    X = country_data['Year'].values.reshape(-1, 1)
    y = country_data['Happiness Score'].values
    model = LinearRegression()
    model.fit(X, y)
    
    # Predict happiness score for 2030
    future_score = model.predict(np.array([[2030]]))
    predictions.append(future_score[0])
```

```python
# Create a DataFrame for predictions
predicted_data = pd.DataFrame({
    'Country': top_countries['Country'],
    'Predicted Happiness Score (2030)': predictions
})
```

```python
# Visualize the predictions
plt.figure(figsize=(12, 6))
sns.lineplot(data=predicted_data, x='Country', y='Predicted Happiness Score (2030)', marker='o', color='skyblue', linewidth=2.5)

# Add titles and labels
plt.title('Predicted Happiness Scores of Top 10 Happiest Countries by 2030', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Predicted Happiness Score', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
